#Data Visualisation








```python
import matplotlib.pyplot as plt
```

```python
#Bar graph for no of movies with different ratings
df1=df[df.type=='movie']
x=['U','U/A 7+','U/A 13+','U/A 16+','A'] #X-Axis consists of different ratings
y=[]#Y-Axis consists of number of films having that particular rating
for i in x:
    t=df1[df1.age_rating==i]
    r,c=t.shape
    y.append(r)
plt.xlabel('Age_rating')
plt.ylabel('No of movies')
plt.title('Bar graph:No of Movies vs Age ratings')
plt.bar(x,y,label='Hotstar data',color='violet')
plt.legend()
plt.show()
```

```python
#pie chart on movies with different genres by Highlighting romance
genres=['Action','Animation','Kids','Romance','Thriller','Comedy','Drama','Horror']
movies=[]
for i in genres:
    t=df1[df1.genre==i]
    r,c=t.shape
    movies.append(r)
print(movies)
```

```python
cols=['green','red','brown','cyan','magenta','blue','orange','violet']
plt.pie(movies,labels=genres,colors=cols,startangle=90,explode=(0,0,0,0.2,0,0,0,0),shadow=True,autopct='%.1f%%')
plt.title('Hotstar_movies_genre')
plt.show()
```

```python
tv_shows=df[df.type=='tv']
x=[2012,2013,2014,2015,2016,2017,2018,2019,2020]
y=[]
for i in x:
  t=tv_shows[tv_shows.year==i]
  r,c=t.shape
  y.append(r)
print(y)
```

```python
#Line graph to find no of tv shows released in hotstar from 2012 t0 2020
plt.plot(x,y,'purple')
plt.title('No of TV shows released from 2012 to 2020')
plt.xlabel('Years')
plt.ylabel('No of TV Shows released')
plt.show()
```

```python
seasons=tv_shows['seasons'].tolist()
bins=[0,5,10,15,20,25,30]
plt.hist(seasons,bins,histtype='bar',rwidth=0.8,color='cyan')
plt.xlabel('No of seasons')
plt.ylabel('TV Shows released')
plt.title('Histogram of seasons of the TV Shows')
plt.legend()
plt.show()
```
